```python
import pandas as pd
import numpy as np

mhs = pd.read_excel("Data Kelulusan Mahasiswa 2022.xlsx")
mhs.head(10)




```

```python
x = mhs.drop(["StatusKelulusan"], axis=1)
x.head(5)
```

```python
y = mhs["StatusKelulusan"]
y.head(10)
```

```python
#import Gaussian Naive Byaes model
from sklearn.naive_bayes import GaussianNB
```

```python
# Memamnggil fungsi Gaussian Naive bayes
nbc = GaussianNB()
```

```python
# memasukkan data trainiing pada fungsi klasifikasi naive bayes
data_training = nbc.fit(x,y)
```

```python
# melakukan prediksi pada data training
y_predict = data_training.predict(x)
print(y_predict)
```

```python
# Melakukan prediksi pada data input => data yang akan di prediksi
Jenis_Kelamin = 1  #pria
Status_Kerja = 0 #mahasiswa
Umur = 25
Status_Nikah = 1 #Menikah
IPS1 = 3.12
IPS2 = 2.75
IPS3 = 2.05
IPS4 = 3.01
IPS5 = 2.56
IPS6 = 3.67
IPS7 = 2.75
IPS8 = 3.45
IPK = np.average([IPS1, IPS2, IPS3, IPS4, IPS5, IPS6, IPS7, IPS8])
Data_Testing = [[Jenis_Kelamin, Status_Kerja, Umur, Status_Nikah, IPS1, IPS2, IPS3, IPS4, IPS5, IPS6, IPS7, IPS8, IPK ]]
```

```python
# Melakukan prediksi berdasarkan data input
y_pred = data_training.predict(Data_Testing)
```

```python
# Mencetak hasil prediksi kelulusan mahasiswa
# Hasil 0 => Terlambat
#Hasil 1 => Tepat

if y_pred == 0:
  hasil = "Terlambat"
elif y_pred == 1:
  hasil = "Tepat"
else:
  hasil = "Eror"
print("Hasil prediksi kelulusan mahasiswa =",hasil)
```

```python
# Menghitung Akurasi => Melihat Skor Akurasi
from sklearn.metrics import accuracy_score
print("Nilai Akurasi = %0.2f" % accuracy_score(y,y_predict))
```

```python
# Mencetak Laporan Hasil Klasifikasi
from sklearn.metrics import classification_report
print(classification_report(y,y_predict))
```
